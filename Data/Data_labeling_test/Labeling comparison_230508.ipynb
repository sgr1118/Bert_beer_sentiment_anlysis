```python
# 통합된 라벨링 csv 파일 불러오기

import pandas as pd

df = pd.read_csv('self_alpaca_gpt_vader.csv')

print(len(df))
```

```python
print(df['label'].value_counts())
print(df['alpack_label'].value_counts())
print(df['gpt_label'].value_counts())
print(df['vader_label'].value_counts())
```

```python
# self-label과 일치 여부 확인하기

df['self-alpack'] = df['label'] == df['alpack_label']
df['self-gpt'] = df['label'] == df['gpt_label']
df['self-vader'] = df['label'] == df['vader_label']
```
